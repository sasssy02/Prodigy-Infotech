```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('/content/bank-additional-full.csv',sep=';')
```

```python
data.head()
```

```python
# data statistics
data.describe()
```

```python
# Create a histogram for the 'age' column
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Ages in the Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a bar chart for the 'job' column
job_counts = data['job'].value_counts()

plt.figure(figsize=(12, 6))
job_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Distribution of Job Categories in the Dataset')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# 1. Bar Chart for Education Levels

```python
# Create a bar chart for the 'education' column
education_counts = data['education'].value_counts()

plt.figure(figsize=(12, 6))
education_counts.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# 2. Box Plot for age by marital Status

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='marital', y='age', data=data, palette='pastel')
plt.title('Age Distribution by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Age')
plt.show()
```

# 3. Pie Chart for Contact Method Distribution

```python
contact_counts = data['contact'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(contact_counts, labels=contact_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Contact Methods')
plt.show()
```

# 4. Scatter Plot to Show Relationship Between age and duration

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['duration'], alpha=0.5, color='purple')
plt.title('Relationship Between Age and Duration of Call')
plt.xlabel('Age')
plt.ylabel('Duration (seconds)')
plt.show()
```

# 5. Histogram for duration of Calls

If you want to visualize the distribution of the duration of calls, use this code:

```python
plt.figure(figsize=(10, 6))
plt.hist(data['duration'], bins=30, color='green', edgecolor='black')
plt.title('Distribution of Call Duration')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.show()
```

# 6. Count Plot for y (Outcome of Campaign)

This shows the number of clients who subscribed or did not subscribe to the product:

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='y', data=data, palette='viridis')
plt.title('Outcome of the Campaign (Subscription)')
plt.xlabel('Subscribed')
plt.ylabel('Count')
plt.show()
```

# 7. Pair Plot to See Relationships Across Multiple Variables

```python
sns.pairplot(data[['age', 'duration', 'campaign']], diag_kind='kde', plot_kws={'alpha':0.6})
plt.suptitle('Pair Plot of Age, Duration, and Campaign', y=1.02)
plt.show()
```
